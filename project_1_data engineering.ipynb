```python
def currency_data(id, vs_currency):
    data = cg.get_coin_ohlc_by_id(id, vs_currency, days = "30")
    df = pd.DataFrame(data)
    df.columns = ["Date", "Open", "High", "Low", "Close"]
    df['Coin']= id
    df['Curr']= vs_currency
    df['Date'] = pd.to_datetime(df['Date'],unit='ms')
    
    return df
```
